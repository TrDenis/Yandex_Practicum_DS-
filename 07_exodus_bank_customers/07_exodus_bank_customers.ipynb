# Отток клиентов

## Введение

Из "Бета-Банка" стали уходить клиенты. Каждый месяц. Немного, но заметно. Банковские маркетологи посчитали: сохранять текущих клиентов дешевле, чем привлекать новых.

В нашем распоряжении исторические данные о поведении клиентов и расторжении договоров с банком.

**Задача: спрогнозировать, уйдёт клиент из банка в ближайшее время или нет.**

Построим модель с предельно большим значением *F1-меры*, не менее 0.59. Проверим *F1-меру* на тестовой выборке. Для построенной модели измерим *AUC-ROC*, и сравним её значение с *F1-мерой*.

Источник данных: [https://www.kaggle.com/barelydedicated/bank-customer-churn-modeling](https://www.kaggle.com/barelydedicated/bank-customer-churn-modeling)

## Описание данных

**Признаки:**

*RowNumber* — индекс строки в данных
*CustomerId* — уникальный идентификатор клиента
*Surname* — фамилия
*CreditScore* — кредитный рейтинг
*Geography* — страна проживания
*Gender* — пол
*Age* — возраст
*Tenure* — количество недвижимости у клиента
*Balance* — баланс на счёте
*NumOfProducts* — количество продуктов банка, используемых клиентом
*HasCrCard* — наличие кредитной карты
*IsActiveMember* — активность клиента
*EstimatedSalary* — предполагаемая зарплата

**Целевой признак:**

*Exited* — факт ухода клиента

Рассмотрим гистограмму распределения значений, возможно, она покажет что это за пропуски. Также выведем 10 случайных строк, содержащих пропуски и посмотрим на данные в других столбцах, попробуем понять взаимосвязь.

```python
# посчитаем количество пропусков
print('Количество пропусков:', df['tenure'].isna().sum())

# построим гистограмму распределения значений
plt.figure(figsize=(10, 7))
sns.histplot(data=df['tenure'].value_counts(), x=df['tenure'], color='skyblue', bins=11)
plt.title('Гистограмма распределения значений столбца "tenure"', weight = 'bold')
plt.xlabel('Количество лет', color='grey', weight = 'bold')
plt.ylabel('Количество клиентов', color='grey', weight = 'bold')
plt.show()

# выведем 10 случайных строк с пропусками
display(df.loc[df['tenure'].isna() == True].sample(10))
```

Чёткой и понятной причины появления пропусков определить не получается, также не видно взаимосвязи пропусков с какими-либо другими данными.
Количество пропусков приблизительно равно количеству значений, которых "не хватает" в категориях 0 и 10 лет. Заполним случайным образом пропуски значениями 0 и 10 и вновь рассмотрим гистограмму распределения значений.

```python
# рандомно заполним пропуски значениями 0 и 10
np.random.seed(MY_SEED)
df.loc[df['tenure'].isna(), 'tenure'] = np.random.choice([0, 10], size=df['tenure'].isna().sum())

# посчитаем количество пропусков
print('Количество пропусков:', df['tenure'].isna().sum())

# построим гистограмму распределения значений
plt.figure(figsize=(10, 7))
sns.histplot(data=df['tenure'].value_counts(), x=df['tenure'], color='skyblue', bins=11)
plt.title('Гистограмма распределения значений столбца "tenure"', weight = 'bold')
plt.xlabel('Количество лет', color='grey', weight = 'bold')
plt.ylabel('Количество клиентов', color='grey', weight = 'bold')
plt.show()
```

Пропуски заполнены, значения распределились более-менее равномерно.

## Исследование задачи

### Разбиение данных на выборки

Создадим признаки для создания обучающей модели и выделим из данных целевой признак. Разделим данные на три выборки: обучающую, валидационную и тестовую.

```python
# выделим целевой признак
target = df['exited']
# создадим признаки для модели
features = df.drop('exited', axis=1)

# отделим обучающий (train) и тестовый (test) датасеты
features_train, features_test, target_train, target_test = train_test_split(
    features, target, test_size=0.4, random_state=MY_SEED, stratify=target)

# разделим датасеты на тестовый (test) и валидационный (valid)
features_valid, features_test, target_valid, target_test = train_test_split(
    features_test, target_test, test_size=0.5, random_state=MY_SEED, stratify=target_test)

# выведем сведения о разбивке данных
print(f'Данные разбиты в соотношении: '
      f'{len(features_train) / len(features):.1f}:'
      f'{len(features_valid) / len(features):.1f}:'
      f'{len(features_test) / len(features):.1f}')
```

В результате получили три набора данных:
1. Обучающий (на нём будем производить обучение моделей)
2. Валидационный (используется для проверки качества модели)
3. Тестовый (для оценки готовой модели)

### Прямое кодирование

Преобразуем категоральные признаки столбцов `'geography'` и `'gender'` методом *One-Hot Encoding*. Мужчин закодируем `'1'`, женщин `'0'`.

```python
# создадим списки признаков категоральных и количественных
cat = list(features_train.select_dtypes('object').columns)
num = list(features_train.select_dtypes(exclude='object').columns)

# масштабируем
scaler = StandardScaler()
features_train_scaled = pd.DataFrame(scaler.fit_transform(features_train[num]), columns=[num], index=features_train.index)
features_valid_scaled = pd.DataFrame(scaler.fit_transform(features_valid[num]), columns=[num], index=features_valid.index)
features_test_scaled = pd.DataFrame(scaler.transform(features_test[num]), columns=[num], index=features_test.index)

# кодируем методом OHE
enc = OneHotEncoder(handle_unknown='ignore')

features_train_ohe = enc.fit_transform(features_train[cat])
features_valid_ohe = enc.fit_transform(features_valid[cat])
features_test_ohe = enc.transform(features_test[cat])

# объединяем количественные и категоральные признаки в единую выборку
features_train_lr = pd.concat([features_train_scaled, pd.DataFrame(features_train_ohe.toarray(), index=features_train.index)], axis=1)
features_valid_lr = pd.concat([features_valid_scaled, pd.DataFrame(features_valid_ohe.toarray(), index=features_valid.index)], axis=1)
features_test_lr = pd.concat([features_test_scaled, pd.DataFrame(features_test_ohe.toarray(), index=features_test.index)], axis=1)

# проверим результаты преобразований, выведем пять случайных строк датафреймов
display(features_train_lr.sample(5))
display(features_valid_lr.sample(5))
display(features_test_lr.sample(5))
```

Теперь можно обучать, проверять, тестировать и исследовать модели. Выберем в качестве базовой модель *"Случайный лес"*, потому что она не самая простая, интересная и обычно позволяет достичь хороших результатов и рассчитаем для неё метрику *F1-меры*.

### Создание и оценка модели "Случайный лес"

```python
# создадим объект для классификации с заданным количеством деревьев и глубины дерева
model = RandomForestClassifier(random_state=MY_SEED, n_estimators=100, max_depth=11)

# обучим модель на основе тренировочный выборки
model.fit(features_train_lr, target_train)

# получим предсказание модели на валидационной выборке
predicted_valid = model.predict(features_valid_lr)

# рассчитаем значения F1-меры
f1 = f1_score(target_valid, predicted_valid)

# выведем значение F1-меры
print(f'F1-мера = {f1:.4f}')
# создадим датафрейм с результатами тестирования
df_results = pd.DataFrame({'test_name': ['Первая оценочная модель RandomForest'], 'F1-мера': [f1],
                           'Время обучения (минуты)': [0], 'params': ['n_estimators=100, max_depth=11']})
```

Как видим, интересующая метрика не "дотягивает" до заданного значения. Попробуем увеличить значение *F1-меры*, для этого дополнительно исследуем данные на наличие дисбаланса и подберём оптимальные гиперпараметры для нашей модели.

## Борьба с дисбалансом

### Подготовка данных и параметров для моделей

```python
# объединим тренировочный и валидационный датасеты для использования кросс-валидации
features_train_lr = pd.concat([features_train_lr, features_valid_lr])
target_train = pd.concat([target_train, target_valid])

# зададим диапазоны гиперпараметров для подбора
param_distributions = {'n_estimators': range(50, 325, 25),
                       'max_depth': range(1, 18, 1),
                       'min_samples_split': range(2, 52, 5),
                       'min_samples_leaf': range(2, 52, 5)}

# подготовим кросс-валидацию на 5-ти выборках
kf = StratifiedKFold(n_splits=7, shuffle=True, random_state=MY_SEED)
```

### Изучение методов балансировки классов

В ходе этого исследования изучим модели *"Случайного леса"* в сочетании с различными методами балансировки классов. Рассмотрим полученные результаты метрики *F1-меры*, а также затраченное время (в минутах) на обучение. Дополнительно рассмотрим модель *"Градиентного бустинга"*. Кроме того, будем использовать масштабирование данных с целью добиться наилучшего значения *F1-меры*.

#### Параметр *class_weight* алгоритма *Random Forest*

```python
# создадим классификатор модели случайного леса с параметром балансировки классов
rfc = RandomForestClassifier(class_weight='balanced', random_state=MY_SEED)

# создадим объект модели на основе классификатора, диапазонов гиперпараметров,
# с количеством переборов 200, с выводом минимальных сообщений, максимальной загрузкой ядер процессора
rs = RandomizedSearchCV(rfc, param_distributions, n_iter=200, verbose=1, cv=kf,
                        n_jobs=-1, scoring='f1', return_train_score=True, random_state=MY_SEED)

# зафитим модель и замерим время моделирования
start_time = time.time()
rs.fit(features_train_lr, target_train)
run_time = (time.time() - start_time) / 60

# выведем значение F1-меры
print(f'F1-мера = {rs.best_score_:.4f}')

# добавим результаты тестирования в датасет
df_results.loc[len(df_results.index)] = ['Модель RF с параметром балансировки классов и подбором параметров',
                                         rs.best_score_, run_time, rs.best_params_]
```

#### *StandardScaler* + *SMOTE* + *Random Forest*

Снова подготовим данные, уже для обработки в пайплайне.

```python
# выделим целевой признак
target = df['exited']
# создадим признаки для модели
features = df.drop('exited', axis=1)

# отделим обучающий (train) и тестовый (test) датасеты
features_train, features_test, target_train, target_test = train_test_split(
    features, target, test_size=0.2, random_state=MY_SEED, stratify=target)

# для пайплайна добавим 'RFC__' к названию каждого гиперпараметра
new_params = {'RFC__' + key: param_distributions[key] for key in param_distributions}
```

Обучим модель.

```python
# создадим шаги для обработки
# масштабирование и категоризация признаков (FPP - features pre-processing)
step_FPP = ('FPP', ColumnTransformer([
    ('num', make_pipeline(StandardScaler()),
     make_column_selector(dtype_include=['number'])),
    ('cat', make_pipeline(OneHotEncoder(sparse=False, handle_unknown='ignore')),
     make_column_selector(dtype_exclude=['number']))]))
# балансировка классов (TPP - target pre-processing)
step_TPP = ('TPP', SMOTE(random_state=MY_SEED))
# классификатор модели
step_RFC = ('RFC', RandomForestClassifier(random_state=MY_SEED))

# зададим последовательность обработки
pipeline_RFC = Pipeline(steps=[step_FPP, step_TPP, step_RFC])

# создадим объект модели
rs_imba = RandomizedSearchCV(pipeline_RFC, new_params, n_iter=200, cv=kf, verbose=1,
                             n_jobs=-1, scoring='f1', return_train_score=True, random_state=MY_SEED)

# зафитим модель и замерим время моделирования
start_time = time.time()
rs_imba.fit(features_train, target_train)
run_time = (time.time() - start_time) / 60

# выведем значение F1-меры
print(f'F1-мера = {rs_imba.best_score_:.4f}')

# добавим результаты тестирования в датасет
df_results.loc[len(df_results.index)] = ['Модель RF с балансировкой SMOTE и подбором параметров',
                                         rs_imba.best_score_, run_time, rs_imba.best_params_]
```

#### *StandartScaler* + *Near Miss* + *Random Forest*

```python
# создадим шаги для обработки
# масштабирование и категоризация признаков (FPP - features pre-processing)
step_FPP = ('FPP', ColumnTransformer([
    ('num', make_pipeline(StandardScaler()),
     make_column_selector(dtype_include=['number'])),
    ('cat', make_pipeline(OneHotEncoder(sparse=False, handle_unknown='ignore')),
     make_column_selector(dtype_exclude=['number']))]))
# балансировка классов (TPP - target pre-processing)
step_TPP = ('TPP', NearMiss())
# классификатор модели
step_RFC = ('RFC', RandomForestClassifier(random_state=MY_SEED))

# зададим последовательность обработки
pipeline_RFC = Pipeline(steps=[step_FPP, step_TPP, step_RFC])

# создадим объект модели
rs_imba = RandomizedSearchCV(pipeline_RFC, new_params, n_iter=200, cv=kf, verbose=1,
                             n_jobs=-1, scoring='f1', return_train_score=True, random_state=MY_SEED)

# зафитим модель и замерим время моделирования
start_time = time.time()
rs_imba.fit(features_train, target_train)
run_time = (time.time() - start_time) / 60

# выведем значение F1-меры
print(f'F1-мера = {rs_imba.best_score_:.4f}')

# добавим результаты тестирования в датасет
df_results.loc[len(df_results.index)] = ['Модель RF с балансировкой NearMiss и подбором параметров',
                                         rs_imba.best_score_, run_time, rs_imba.best_params_]
```

#### *MinMaxScaler* + *SMOTE* + *Gradient Boosting*

```python
# создадим шаги для обработки
# масштабирование и категоризация признаков (FPP - features pre-processing)
step_FPP = ('FPP', ColumnTransformer([
    ('num', make_pipeline(MinMaxScaler()),
     make_column_selector(dtype_include=['number'])),
    ('cat', make_pipeline(OneHotEncoder(sparse=False, handle_unknown='ignore')),
     make_column_selector(dtype_exclude=['number']))]))
# балансировка классов (TPP - target pre-processing)
step_TPP = ('TPP', SMOTE(random_state=MY_SEED))
# классификатор модели
step_RFC = ('RFC', GradientBoostingClassifier(random_state=MY_SEED))

# зададим последовательность обработки
pipeline_RFC = Pipeline(steps=[step_FPP, step_TPP, step_RFC])

# создадим объект модели
rs_imba = RandomizedSearchCV(pipeline_RFC, new_params, n_iter=200, cv=kf, verbose=1,
                             n_jobs=-1, scoring='f1', return_train_score=True, random_state=MY_SEED)

# зафитим модель и замерим время моделирования
start_time = time.time()
rs_imba.fit(features_train, target_train)
run_time = (time.time() - start_time) / 60

# выведем значение F1-меры
print(f'F1-мера = {rs_imba.best_score_:.4f}')

# добавим результаты тестирования в датасет
df_results.loc[len(df_results.index)] = ['Модель GB с балансировкой SMOTE и подбором параметров',
                                         rs_imba.best_score_, run_time, rs_imba.best_params_]
```

#### Итоги исследования методов балансировки классов

```python
# выведем датафрейм с полученными результатами
pd.set_option('display.float_format', '{:,.4f}'.format)
results = pd.DataFrame(rs_imba.cv_results_)
display(df_results)
```

Балансировка классов дала хорошее улучшение метрик, кроме метода *"NearMiss()"*. Самый лучший результат метрики *F1-мера* был получен на модели *Gradient Boosting* с масштабированием *MinMaxScaler* и балансировкой классов методом *SMOTE*. Однако, хотелось бы заметить, что времени на обучение этой модели понадобилось более **50 минут**, и если сравнить эту модель с моделью *Random Forest* и методом балансировки самого классификатора, то видно, что значение *F1* выросло всего на **0,002**, а времени понадобилось почти в **7 раз больше**. Так же необходимо отметить, что все исследуемые методы находились в "равных условиях" - обучение проводилось на 5-ти выборках с 2000 различными настройками параметров для каждой выборки, т.е. каждый раз обучалось по 1400 моделей с различными параметрами для поиска наиболее оптимальных гиперпараметров.

## Тестирование модели

Произведём предсказание нашей лучшей модели на тестовой выборке. Оценим результаты тестирования по полученным значениям *F1-меры* и *ROC-AUC*.

```python
# получим предсказание модели на тестовой выборке
predicted_test = rs_imba.best_estimator_.predict(features_test)

# получим оценки вероятности положительного исхода
rf_probs = rs_imba.best_estimator_.predict_proba(features_test)[:, 1]

# выведем значение F1-меры и значение ROC-AUC для положительного исхода
print(f'F1-мера = {f1_score(target_test, predicted_test):.4f}')
print(f'ROC-AUC = {roc_auc_score(target_test, rf_probs):.4f}')

# выведем значения гиперпараметров модели
display(rs_imba.best_params_)
```

Построим кривую *ROC-AUC* и сравним её с моделью, которая случайно выбирает класс (на графике будет как прямая линия).

```python
# выведем ROC-кривую
fpr, tpr, thresholds = roc_curve(target_test, rf_probs)
plt.figure(figsize=(15, 10))
plt.plot(fpr, tpr, label='Gradient Boosting')

# ROC-кривая случайной модели (выглядит как прямая)
plt.plot([0, 1], [0, 1], linestyle='--', label='RandomModel')

# установим границы осей от 0 до 1
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])

plt.xlabel('False Positive Rate', color='grey', weight = 'bold')
plt.ylabel('True Positive Rate', color='grey', weight = 'bold')
plt.title('ROC-кривая', weight = 'bold')
plt.legend()
plt.show()
```

Мы получили хорошую модель, которая показала на тестовой выборке результаты лучше, чем во время обучения. Также видно, что модель значительно лучше предсказывает, чем при использовании случайного предсказания - достаточно сравнить площади под соответствующими графиками: **0,5** для случайной и **0,8717** для нашей модели.
